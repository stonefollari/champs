### Merging Structure data into train_df

```python
# Merge structure data onto train_df for each atom (atom_index_0 and atom_index_1). Hold in mol_df
mol_df = train_df.merge(structures_df, left_on=['molecule_name','atom_index_0'], right_on=['molecule_name','atom_index'])
mol_df = mol_df.merge(structures_df, left_on=['molecule_name','atom_index_1'], right_on=['molecule_name','atom_index'])

# drop extra columns from merge and rename
mol_df.drop(['atom_index_x','atom_index_y'], axis=1, inplace=True)
mol_df.rename(columns={'atom_x':'atom_0','atom_y':'atom_1','x_x':'x_0','y_x':'y_0','z_x':'z_0','x_y':'x_1','y_y':'y_1','z_y':'z_1'}, inplace=True)
```

```python
mol_df
```

# Exploration 1
* Linear Regression of Bond Type & Magnitude of Bond Distance.

### The different kinds of bonds present in the data

```python
def atom_displacement(row):
    return np.linalg.norm(np.array([row['x_1']-row['x_0'],row['y_1']-row['y_0'],row['z_1']-row['z_0']]))
```

```python
mol_df['displacement'] = mol_df.apply(lambda row: atom_displacement(row), axis=1)
```

```python
mol_df
```

```python
print( mol_df.type.unique() )
```

# Linear Regression
- [Scikit Learn Linear Regression](https://scikit-learn.org/stable/modules/linear_model.html)

```python
type1 = mol_df[ mol_df.type == '1JHC'][['scalar_coupling_constant','displacement']]
```

```python
type1_x = type1['displacement'].values
type1_X = type1_x.reshape(-1,1)
type1_y = type1['scalar_coupling_constant'].values
```

```python
reg = linear_model.LinearRegression()
reg.fit( type1_x, type1_y )
```

```python
print(reg.coef_)
print(reg.intercept_)
```

```python
num = 5000
x = type1_x[:num]
y = type1_y[:num]

plt.scatter(x, y)
plt.plot(x, x*reg.coef_ + reg.intercept_, c='r', linewidth=1, linestyle='--', dashes=[1,1])
plt.xlim([1.06,1.13])
plt.ylim([70,150])
plt.title("Linear Regression Fit of Scalar-Coupling-Constant vs. Bond Displacement")
plt.xlabel('Displacement')
plt.ylabel('Scalar Coupling Constant')
plt.show()
```
